```python
import pandas as pd
```

```python
df = pd.read_csv('estadisticas_partidos.csv', sep=',')
```

```python
df_home = df[['team_home'] + [col for col in df.columns if 'home' in col]]
df_away = df[['team_away'] + [col for col in df.columns if 'away' in col]]
```

```python
df_home.columns = [col.replace('_home', '') for col in df_home.columns]
df_away.columns = [col.replace('_away', '') for col in df_away.columns]
```

```python
df_home = df_home.loc[:,~df_home.columns.duplicated()]
df_away = df_away.loc[:,~df_away.columns.duplicated()]
```

```python
df_combined = pd.concat([df_home, df_away])
```

```python
df_combined.groupby('team').sum()
```
